```python
# Another correct way using the columns keyword argument explicitly
df = df.drop(columns=[ 'Country Code', 'City', 'Address',
       'Locality', 'Locality Verbose', 'Longitude', 'Latitude',
       'Currency', 'Has Table booking',
       'Has Online delivery', 'Is delivering now', 'Switch to order menu',
        'Rating color', 'Rating text',
       'Votes'], axis=1)
```

```python
df
```

```python
# Generate dummy variables for 'Cuisines'
cuisines_dummies = pd.get_dummies(df['Cuisines'], drop_first=True)

# Concatenate the dummy variables to the original DataFrame
df = pd.concat([df, cuisines_dummies], axis=1)

# Drop the original 'Cuisines' column if it is no longer needed
df.drop('Cuisines', axis=1, inplace=True)
```

```python
df.columns
```

```python
X = df[['Average Cost for two','Restaurant ID', 'Price range', 'Aggregate rating'] + list(cuisines_dummies.columns)]
y=df[['Restaurant Name']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
st_x= StandardScaler()
x_train= st_x.fit_transform(x_train)
x_test= st_x.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```

```python
# Predict on the test set
y_test_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_test_pred)
print("Accuracy:", accuracy)
```

```python
# Print confusion matrix
conf_matrix = confusion_matrix(y_test, y_test_pred)
print("Confusion Matrix:\n", conf_matrix)
```
